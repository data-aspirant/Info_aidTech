```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#reading the file
data = pd.read_csv("IRIS Info AidTech.csv")
```

```python
print(data)
```

```python
#plotting the dimension of sepal
plt.figure(figsize=(10,5), edgecolor='black')
plt.style.use('default')

plt.scatter(data[data['species']=='Iris-setosa'].sepal_length, data[data['species']=='Iris-setosa'].sepal_width, marker='+', color='navy', edgecolor='black', label='Setosa')
plt.scatter(data[data['species']=='Iris-versicolor'].sepal_length, data[data['species']=='Iris-versicolor'].sepal_width, marker='o', color='lightblue', edgecolor='black', alpha=0.5, label='versicolor')
plt.scatter(data[data['species']=='Iris-virginica'].sepal_length, data[data['species']=='Iris-virginica'].sepal_width, marker='s', color='black', edgecolor='black', label='virginica')

plt.title('Sepal dimensions for each species', fontsize=12, fontweight='bold', backgroundcolor='red')
plt.xlabel('Sepal_length', fontsize=10)
plt.ylabel('Sepal_width', fontsize=10)

plt.grid(linestyle='dotted', lw=.5, color='gray')
plt.legend()
```

```python
#plotting the dimension of petals
plt.figure(figsize=(10,5), edgecolor='black')
plt.style.use('default')

plt.scatter(data[data['species']=='Iris-setosa'].petal_length, data[data['species']=='Iris-setosa'].petal_width, marker='+', color='navy', edgecolor='black', label='Setosa')
plt.scatter(data[data['species']=='Iris-versicolor'].petal_length, data[data['species']=='Iris-versicolor'].petal_width, marker='o', color='lightblue', edgecolor='black', alpha=0.5, label='versicolor')
plt.scatter(data[data['species']=='Iris-virginica'].petal_length, data[data['species']=='Iris-virginica'].petal_width, marker='s', color='black', edgecolor='black', label='virginica')

plt.title('petal dimensions for each species', fontsize=12, fontweight='bold', backgroundcolor='lightgreen')
plt.xlabel('petal_length', fontsize=10)
plt.ylabel('petal_width', fontsize=10)

plt.grid(linestyle='dotted', lw=.5, color='gray')
plt.legend()
```

```python
#seperating dependent and independent variables
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
#converting dependent variable into numberical value
from sklearn.preprocessing import LabelEncoder
lc = LabelEncoder()
y=lc.fit_transform(y)
```

```python
print(y)
```

```python
#seperating the data into training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#making a RandonForestClassifier model and training it using training data
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(x_train,y_train)
```

```python
#predicting the dependent variable
y_pred = classifier.predict(x_test)
#printing the predected and test variable
np.printoptions(precision=2)
print(np.concatenate((y_pred.reshape(-1,1),y_test.reshape(-1,1)),1))
```

```python
#checking confusion matrix and accuracy score
from sklearn.metrics import accuracy_score,confusion_matrix
cm = confusion_matrix(y_pred,y_test)
print(cm)
print("{}%".format(accuracy_score(y_pred,y_test)*100))
```

```python
#performing cross validation
from sklearn.model_selection import cross_val_score
cv = cross_val_score(estimator=classifier,X=x_test,y=y_test,cv=10)
print("Accuracy: {:.2f} %".format(cv.mean()*100))
print("Standard Deviation: {:.2f} %".format(cv.std()*100))
```
